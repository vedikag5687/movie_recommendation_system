```python
import numpy as np
import pandas as pd
import ast
import nltk
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
credit = pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credit.head()
```

```python
new = movies.merge(credit, on = 'title')
```

```python
new.head(1)
```

```python
new.shape
```

```python
new.info()
```

```python
movie = new[['movie_id', 'genres', 'keywords', 'overview', 'title', 'cast', 'crew']]
```

```python
movie.head()
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.duplicated().sum()
```

```python
movie.iloc[0].genres
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i ['name'])
    return l
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
           l.append(i ['name'])
           counter+1
        else:
            break
    return l
```

```python
movie['cast'] = movie['cast'].apply(convert3)
```

```python
movie.head()
```

```python
movie['crew'][0]
```

```python
import ast

def fetch_director(obj):
    l = []
    try:
        crew_list = ast.literal_eval(obj)
        for i in crew_list:
            if i.get('job') == 'Director':
                l.append(i.get('name'))
                break
    except (SyntaxError, ValueError):
        pass

    return l
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie.head()
```

```python
movie['overview'][0]
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ", "")for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ", "")for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ", "")for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
movie.head()
```

```python
movie['tags'] = movie['genres'] + movie['cast'] + movie['crew'] + movie['keywords'] + movie['overview']
```

```python
movie.head()
```

```python
new_df = movie[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
feature_names = cv.get_feature_names_out()

print(feature_names.tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Avatar')
```
